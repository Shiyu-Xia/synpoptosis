```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
file_path_a1 = './export_01-Well-A1_ECD-A+.csv'
file_path_a2 = './export_01-Well-A2_ECD-A+.csv'
file_path_b1 = './export_01-Well-B1_ECD-A+.csv'
file_path_b2 = './export_01-Well-B2_ECD-A+.csv'
file_path_c1 = './export_01-Well-C1_ECD-A+.csv'
file_path_c2 = './export_01-Well-C2_ECD-A+.csv'
data_a1 = pd.read_csv(file_path_a1)
data_a2 = pd.read_csv(file_path_a2)
data_b1 = pd.read_csv(file_path_b1)
data_b2 = pd.read_csv(file_path_b2)
data_c1 = pd.read_csv(file_path_c1)
data_c2 = pd.read_csv(file_path_c2)
ecd_a1, pb450_a1 = data_a1['ECD-A'], data_a1['PB450-A']
ecd_a2, pb450_a2 = data_a2['ECD-A'], data_a2['PB450-A']
ecd_b1, pb450_b1 = data_b1['ECD-A'], data_b1['PB450-A']
ecd_b2, pb450_b2 = data_b2['ECD-A'], data_b2['PB450-A']
ecd_c1, pb450_c1 = data_c1['ECD-A'], data_c1['PB450-A']
ecd_c2, pb450_c2 = data_c2['ECD-A'], data_c2['PB450-A']
def calculate_positive_percentage(data, threshold=3e4):
    positive_count = np.sum(data > threshold)  
    total_count = len(data)  
    percentage = (positive_count / total_count) * 100 if total_count > 0 else 0
    return percentage
bins = np.logspace(np.log10(2e3), np.log10(3e5), 21) 
def calculate_percentages_for_well(ecd_data, pb450_data):
    indices = np.digitize(ecd_data, bins)
    return [calculate_positive_percentage(pb450_data[indices == i]) for i in range(1, len(bins))]
percentages_a1 = calculate_percentages_for_well(ecd_a1, pb450_a1)
percentages_a2 = calculate_percentages_for_well(ecd_a2, pb450_a2)
percentages_b1 = calculate_percentages_for_well(ecd_b1, pb450_b1)
percentages_b2 = calculate_percentages_for_well(ecd_b2, pb450_b2)
percentages_c1 = calculate_percentages_for_well(ecd_c1, pb450_c1)
percentages_c2 = calculate_percentages_for_well(ecd_c2, pb450_c2)
bin_centers = (bins[:-1] + bins[1:]) / 2
plt.figure(figsize=(10, 8))
plt.plot(bin_centers, percentages_a1, marker='o', color='red', label='Cherry only, Replicate 1')
plt.plot(bin_centers, percentages_a2, marker='o', color='blue', label='Autoinhibited Caspase-3, Replicate 1')
plt.plot(bin_centers, percentages_b1, marker='o', color='red', label='Cherry only, Replicate 2')
plt.plot(bin_centers, percentages_b2, marker='o', color='blue', label='Autoinhibited Caspase-3, Replicate 2')
plt.plot(bin_centers, percentages_c1, marker='o', color='red', label='Cherry only, Replicate 3')
plt.plot(bin_centers, percentages_c2, marker='o', color='blue', label='Autoinhibited Caspase-3, Replicate 3')
plt.xscale('log')
plt.xlabel('ECD-A')
plt.ylabel('Percentage of PB450-A-positive Cells')
plt.title('Percentage of PB450-A-positive Cells in Each ECD-A Bin (Separate Wells)')
plt.legend()
plt.grid(False)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
file_path_a1 = './export_01-Well-A1_ECD-A+.csv'
file_path_a2 = './export_01-Well-A2_ECD-A+.csv'
file_path_b1 = './export_01-Well-B1_ECD-A+.csv'
file_path_b2 = './export_01-Well-B2_ECD-A+.csv'
file_path_c1 = './export_01-Well-C1_ECD-A+.csv'
file_path_c2 = './export_01-Well-C2_ECD-A+.csv'
data_a1 = pd.read_csv(file_path_a1)
data_a2 = pd.read_csv(file_path_a2)
data_b1 = pd.read_csv(file_path_b1)
data_b2 = pd.read_csv(file_path_b2)
data_c1 = pd.read_csv(file_path_c1)
data_c2 = pd.read_csv(file_path_c2)
ecd_a1, pb450_a1 = data_a1['ECD-A'], data_a1['PB450-A']
ecd_a2, pb450_a2 = data_a2['ECD-A'], data_a2['PB450-A']
ecd_b1, pb450_b1 = data_b1['ECD-A'], data_b1['PB450-A']
ecd_b2, pb450_b2 = data_b2['ECD-A'], data_b2['PB450-A']
ecd_c1, pb450_c1 = data_c1['ECD-A'], data_c1['PB450-A']
ecd_c2, pb450_c2 = data_c2['ECD-A'], data_c2['PB450-A']
def calculate_positive_percentage(data, threshold=3e4):
    positive_count = np.sum(data > threshold)  
    total_count = len(data)  
    percentage = (positive_count / total_count) * 100 if total_count > 0 else 0
    return percentage
bins = np.logspace(np.log10(2e3), np.log10(4e5), 21)  
def calculate_percentages_for_well(ecd_data, pb450_data):
    indices = np.digitize(ecd_data, bins)
    return [calculate_positive_percentage(pb450_data[indices == i]) for i in range(1, len(bins))]
percentages_a1 = calculate_percentages_for_well(ecd_a1, pb450_a1)
percentages_a2 = calculate_percentages_for_well(ecd_a2, pb450_a2)
percentages_b1 = calculate_percentages_for_well(ecd_b1, pb450_b1)
percentages_b2 = calculate_percentages_for_well(ecd_b2, pb450_b2)
percentages_c1 = calculate_percentages_for_well(ecd_c1, pb450_c1)
percentages_c2 = calculate_percentages_for_well(ecd_c2, pb450_c2)
average_abc1 = np.mean([percentages_a1, percentages_b1, percentages_c1], axis=0)
average_abc2 = np.mean([percentages_a2, percentages_b2, percentages_c2], axis=0)
bin_centers = (bins[:-1] + bins[1:]) / 2
plt.figure(figsize=(10, 8))
plt.plot(bin_centers, percentages_a1, 'o', color='red')
plt.plot(bin_centers, percentages_a2, 'o', color='blue')
plt.plot(bin_centers, percentages_b1, 'o', color='red')
plt.plot(bin_centers, percentages_b2, 'o', color='blue')
plt.plot(bin_centers, percentages_c1, 'o', color='red')
plt.plot(bin_centers, percentages_c2, 'o', color='blue')
plt.plot(bin_centers, average_abc1, color='red', label='Cherry only')
plt.plot(bin_centers, average_abc2, color='blue', label='Autoinhibited caspase-3')
plt.xscale('log')
plt.xlabel('Cherry fluorescence')
plt.ylabel('Percentage of Annexin-positive Cells')
plt.title('Percentage of Annexin-positive Cells in Each Cherry Bin')
plt.legend()
plt.grid(False)
pdf_file_path = './Annexin_positive_percentage_plot.pdf'
plt.savefig(pdf_file_path, format='pdf')
plt.show()
```
